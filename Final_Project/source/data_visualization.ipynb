### 2.1. Câu hỏi 1:

**Câu hỏi:**
- Mối quan hệ giữa số lượng chương (Chapters) và tập (Volumns) với số lượng người theo dõi bộ truyện (Members) và mức độ tương tác của họ (Favourite)?

**Mục tiêu:**
- Phân tích mối quan hệ giữa **số lượng chương**, **số lượng tập**, **số lượng người đọc**, và **số lượng yêu thích** sẽ giúp ta hiểu rõ hơn về cách các yếu tố này ảnh hưởng đến sự phổ biến và mức độ gắn kết của một bộ truyện đối với độc giả. 
- Những phân tích này rất hữu ích cho các nhà xuất bản truyện, các tác giả, và những ai quan tâm đến việc phát triển nội dung truyện một cách hiệu quả và thu hút người đọc lâu dài. và thu hút người đọc lâu dài.

**Ý tưởng:**
- Tính toán và vẽ ma trận tương quan giữa Chapter, Volume, Members và Favorites.  
- Vẽ biểu đồ scatter giữa Chapter, Volume, Members và Favorites.  
- Thêm đường xu hướng để xác định bất kỳ xu hướng hoặc mối tương quan nào.o.

```python
sns.set(rc={"axes.facecolor":"#F2EAC5","figure.facecolor":"#F2EAC5"})
# Chọn ra các cột cần thiết từ dataframe
heatmap_data = manga_df[['Chapters', 'Volumes', 'Members', 'Favorite']]

# Tính toán ma trận tương quan giữa các thuộc tính
correlation_matrix = heatmap_data.corr()

# Vẽ heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Heatmap between Chapters, Volumes, Members, and Favorites')
plt.show()
```

```python
# Hàm để vẽ đường hồi quy cho mỗi cặp thuộc tính
def plot_regression_line(x, y, ax):
    # Xử lý dữ liệu để có thể thực hiện hồi quy
    x_reshape = x.values.reshape(-1, 1)
    model = LinearRegression()
    model.fit(x_reshape, y)
    y_pred = model.predict(x_reshape)
    
    # Vẽ biểu đồ phân tán và đường hồi quy
    ax.scatter(x, y, c ='blue' ,marker='x', label='Data points')
    ax.plot(x, y_pred, color='red', label='Regression line')
    ax.set_xlabel(x.name)
    ax.set_ylabel(y.name)
    ax.legend()

# Vẽ biểu đồ phân tán với đường hồi quy cho các cặp thuộc tính
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Cặp 1: Chapters vs Members
plot_regression_line(manga_df['Chapters'], manga_df['Members'], axs[0, 0])
axs[0, 0].set_title('Chapters vs Members')

# Cặp 2: Chapters vs Favorite
plot_regression_line(manga_df['Chapters'], manga_df['Favorite'], axs[0, 1])
axs[0, 1].set_title('Chapters vs Favorite')

# Cặp 3: Volumes vs Members
plot_regression_line(manga_df['Volumes'], manga_df['Members'], axs[1, 0])
axs[1, 0].set_title('Volumes vs Members')

# Cặp 4: Volumes vs Favorite
plot_regression_line(manga_df['Volumes'], manga_df['Favorite'], axs[1, 1])
axs[1, 1].set_title('Volumes vs Favorite')

plt.tight_layout()
plt.show()
```

**Nhận xét:**

1. **Ma trận tương quan:**
- **Tương quan giữa Members và Favorites**: Một tương quan mạnh mẽ với hệ số 0.88 cho thấy rằng khi truyện có nhiều thành viên hơn, nó thường nhận được nhiều lượt yêu thích hơn, chỉ ra mối liên hệ chặt chẽ giữa quy mô thành viên và mức độ tương tác.
- **Tương quan giữa Chapters và Volumes**: Tương quan rất mạnh với hệ số 0.63, cho thấy rằng truyện có nhiều chương thường có nhiều tập hơn, điều này là hiển nhiên vì các chương thường được tập hợp thành các tập.  
- **Tương quan giữa Chapters/Volumes và Members/Favorites**:  
  - Tương quan dương nhưng yếu, cho thấy có xu hướng tăng nhẹ nhưng không đủ mạnh để dự đoán chính xác.  
  - Điều này có nghĩa là một truyện dài hơn không đảm bảo sẽ có nhiều thành viên hoặc lượt yêu thích hơn.  

2. **Phân tích đường xu hướng:**
- Có xu hướng tăng giữa **Chapters và Members**, cũng như **Volumes và Members**, cho thấy rằng nội dung nhiều hơn thường dẫn đến lượng thành viên lớn hơn.  
- Xu hướng yếu hơn giữa **Chapters và Favorites**, cũng như **Volumes và Favorites**, cho thấy độ dài nội dung có tác động hạn chế đến lượt yêu thích.  
- Sự phân tán trong tất cả các biểu đồ cho thấy sự biến động chưa được giải thích chỉ bằng độ dài nội dung.  
- Điều này chỉ ra rằng các yếu tố khác (như thể loại, danh tiếng tác giả, chất lượng cốt truyện) có ảnh hưởng đến mức độ tương tác và số lượng thành viên.

### 2.2. Câu hỏi 2:

**Câu hỏi:**
- Số lượng truyện được phát hành đã thay đổi như thế nào theo thời gian từ quá khứ đến hiện tại?

**Mục tiêu:**
- Hiểu về sự phát triển lịch sử của truyện: Theo dõi sự phát triển của ngành công nghiệp truyện từ giai đoạn đầu cho đến hiện tại.  
- Xác định xu hướng phát hành các bộ truyện: Quan sát cách số lượng truyện được phát hành thay đổi theo thời gian và nhận diện bất kỳ sự thay đổi hoặc mô hình đáng chú ý nào.  
 
**Ý tưởng:**
- Tạo một cột mới tên là **Year** để trích xuất năm từ cột **Released date**.  
- Tạo một cột mới tên là **Decade** bằng cách chia cột **Year** cho 10 và làm tròn xuống số nguyên gần nhất.
- Nhóm dữ liệu theo **Year** và **Decade**, sau đó tính số lượng truyện phát hành trong mỗi năm và mỗi thập kỷ. 
- Trực quan hóa dữ liệu bằng biểu đồ cột để quan sát xu hướng phát hành truyện theo thời gian.

```python
df_copy = manga_df.copy()
df_copy['Released date'] = pd.to_datetime(df_copy['Released date'], errors='coerce')
# Nhóm dữ liệu theo năm và đếm số lượng truyện được phát hành trong mỗi năm
df_copy['Year'] = df_copy['Released date'].dt.year.astype('Int64')
df_copy['Decade'] = df_copy['Year'] // 10 * 10

manga_per_year = df_copy.groupby('Year').size().reset_index(name='Count')
manga_per_decade = df_copy.groupby('Decade').size().reset_index(name='Count')

# Biểu đồ số lượng truyện được phát hành mỗi năm và số lượng truyện được phát hành mỗi thập kỷ
fig, axes = plt.subplots(1, 2, figsize=(25, 7), gridspec_kw={'width_ratios': [2, 1]})

axes[0].bar(manga_per_year['Year'], manga_per_year['Count'], color="#1d7874")
axes[0].set_title('Number of manga released per year', fontsize=25, weight='bold')
axes[0].set_xlabel('\nYear', fontsize=15, weight='bold')
axes[0].set_ylabel('Number of manga', fontsize=15, weight='bold')
axes[0].set_xticks(range(1950, 2024, 1))
axes[0].set_xticklabels(range(1950, 2024, 1), rotation=90)

axin = axes[0].inset_axes([0.1, 0.4, 0.5, 0.5])
axin.bar(manga_per_year['Year'], manga_per_year['Count'], color="#1d7874", width=0.8)
axin.set_xlim(1949, 1986)
axin.set_ylim(0, 40)
axin.set_xticks(range(1950, 1986, 5))
axin.set_xlabel('Year')
axin.set_ylabel('')

axes[0].indicate_inset_zoom(axin, edgecolor="black", linewidth=1, alpha=0.5)

axes[1].bar(manga_per_decade['Decade'], manga_per_decade['Count'], color="#1d7874", width=8, linewidth=2, edgecolor='k')
axes[1].set_title('Number of manga released per decade', fontsize=25, weight='bold')
axes[1].set_xlabel('Decade', fontsize=15, weight='bold')
axes[1].set_ylabel('Number of manga', fontsize=15, weight='bold')

axin = axes[1].inset_axes([0.1, 0.55, 0.4, 0.4])
axin.bar(manga_per_decade['Decade'], manga_per_decade['Count'], color="#1d7874", width=8, linewidth=2, edgecolor='k')
axin.set_xlim(1955, 1985)
axin.set_ylim(0, 400)
axin.set_xlabel('Decade')
axin.set_ylabel('')
axes[1].indicate_inset_zoom(axin, edgecolor="black", linewidth=1, alpha=0.3)

plt.show()
```

**Nhận xét**

Hai biểu đồ không chỉ giúp chúng ta phân tích số lượng truyện phát hành mà còn cung cấp cái nhìn sâu sắc về sự phát triển của ngành công nghiệp truyện nói chung tại Nhật Bản:

1. **Tổng quan:**
- Dễ dàng nhận thấy rằng số lượng truyện phát hành đã tăng dần qua các năm, nhưng có sự giảm sút đáng kể trong những năm gần đây (từ năm 2020 trở đi). Số lượng truyện phát hành, cũng như quá trình phát triển của ngành công nghiệp truyện Nhật Bản, có thể được chia thành nhiều giai đoạn:

2. **1950-1970:** Giai đoạn đầu
- Mặc dù trước giai đoạn này đã tồn tại các bộ truyện tại Nhật Bản, nhưng dữ liệu chỉ cung cấp thông tin từ năm 1950. Giai đoạn này được coi là khởi đầu của ngành công nghiệp truyện Nhật Bản, với số lượng truyện phát hành tương đối hạn chế, chỉ vài tác phẩm mỗi năm, cao nhất là 6 tác phẩm vào năm 1969.

3. **1970-1985:** Giai đoạn phát triển ban đầu
- Trong giai đoạn này, số lượng truyện phát hành tăng lên so với giai đoạn trước, cho thấy sự khởi đầu của sự phát triển ngành công nghiệp truyện Nhật Bản. Số lượng truyện phát hành tăng dần mỗi năm, khoảng 7-8 tác phẩm mỗi năm. Đây cũng là giai đoạn mà các bộ truyện Nhật Bản, đặc biệt là các bộ manga bắt đầu lan rộng ra toàn cầu với các tác phẩm như *Dragon Ball*, *Tsubasa*, v.v.

4. **1985-2010:** Giai đoạn phát triển mạnh mẽ
- Giai đoạn này chứng kiến sự bùng nổ về số lượng truyện phát hành so với giai đoạn trước (từ 10 tác phẩm mỗi năm lên hơn 200 tác phẩm mỗi năm). Đây cũng là thời kỳ ngành công nghiệp truyện Nhật Bản mở rộng ra toàn cầu với những tác phẩm nổi tiếng như *Naruto*, *One Piece*, *Bleach*, v.v.

5. **2010 đến nay:** Đỉnh cao và sự giảm sút gần đây
- Trong giai đoạn này, ngành công nghiệp truyện Nhật Bản đạt đỉnh cao về số lượng phát hành, độ phổ biến toàn cầu và mức độ bao phủ, với hơn 300 tác phẩm mỗi năm, đạt đỉnh điểm là 356 tác phẩm vào năm 2020. Tuy nhiên, từ năm 2020 trở đi, có thể do tác động nặng nề của đại dịch Covid-19, số lượng truyện phát hành có sự giảm sút.

- Cung cấp thông tin hữu ích cho các nhà xuất bản, người hâm mộ và các chuyên gia trong ngành để lập kế hoạch, chiến lược tiếp thị và ra quyết định một cách hiệu quả

**Ý tưởng:**

- Nhóm dữ liệu theo **Season** và tính tổng số lượng truyện phát hành trong mỗi mùa
- Trực quan hóa dữ liệu bằng biểu đồ cột để quan sát xu hướng và các mô hình phát hành truyện giữa các mùa.   

```python
df_copy['Month'] = df_copy['Released date'].dt.month.astype('Int64')
seasons = ['Spring', 'Summer', 'Fall', 'Winter']
df_copy['Season'] = pd.cut(df_copy['Month'], [0, 3, 6, 9, 12], labels=seasons, include_lowest=True)
# Nhóm 'Decade' và 'Season' với observed=False
manga_per_season = (
    df_copy[df_copy['Year'] >= 1990]
    .groupby(['Decade', 'Season'], observed=False)
    .size()
    .reset_index(name='Count')
)

# Tính phần trăm
manga_per_season['Percentage'] = manga_per_season['Count'] / manga_per_season.groupby('Decade', observed=False)['Count'].transform('sum') * 100

fig, ax = plt.subplots(1, 2, figsize=(20, 8))

# Vẽ biểu đồ area
area_plot_data = manga_per_season.groupby(['Decade', 'Season'], observed=False)['Percentage'].sum().unstack()
area_plot_data.plot(kind='area', stacked=True, ax=ax[0])
ax[0].set_title('Percentage of manga released per season per decade', fontsize=20, weight='bold')
ax[0].set_xlabel('Decade', fontsize=15, weight='bold')
ax[0].set_ylabel('Percentage', fontsize=15, weight='bold')
ax[0].legend(loc='upper left', bbox_to_anchor=(1, 1, 0, 0))
ax[0].set_xticks(range(1990, 2030, 10))

# Thêm số liệu vào biểu đồ area
for decade, row in manga_per_season.groupby('Decade'):
    cumulative_percentage = 0
    for season, percentage in row.set_index('Season')['Percentage'].items():
        cumulative_percentage += percentage
        x_pos = decade
        y_pos = cumulative_percentage
        ax[0].text(x_pos, y_pos - percentage / 2, f"{percentage:.1f}%", ha='center', va='center', fontsize=10)

# Vẽ biểu đồ bar
bar_plot_data = manga_per_season.groupby(['Decade', 'Season'], observed=False)['Count'].sum().unstack()
bar_plot_data.plot(kind='bar', stacked=True, ax=ax[1])
ax[1].set_title('Number of manga released per season per decade', fontsize=20, weight='bold')
ax[1].set_xlabel('Decade', fontsize=15, weight='bold')
ax[1].set_ylabel('Number of manga', fontsize=15, weight='bold')
ax[1].legend(loc='upper left')
ax[1].set_xticklabels(manga_per_season['Decade'].unique(), rotation=0)

# Thêm số liệu vào biểu đồ bar
for bar in ax[1].patches:
    height = bar.get_height()
    if height > 0:
        x_pos = bar.get_x() + bar.get_width() / 2
        y_pos = bar.get_y() + height / 2
        ax[1].text(x_pos, y_pos, f"{int(height):,}", ha='center', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

**Mục tiêu:**

  - Mức độ phổ biến (popularity).

**Ý tưởng:**
- Tách dataset sao cho mỗi thể loại (genre) có một dòng riêng biệt chứa chi tiết về của bộ truyện.  
- Nhóm dữ liệu theo thể loại và tính toán:  
    + Số lượt yêu thích trung bình trên mỗi thể loại.  
    + Số lượng truyện trên mỗi thể loại.  
    + Điểm trung bình trên mỗi thể loại.  
    + Mức độ phổ biến trung bình trên mỗi thể loại.  
- Tạo một điểm tổng hợp (composite score) kết hợp các số liệu này.  
- Xếp hạng các thể loại dựa trên điểm tổng hợp và hiển thị các thể loại hàng đầu. 

```python
manga_data = manga_df.copy()

# Làm sạch và xử lý cột 'Genres'
manga_data['Genres'] = manga_data['Genres'].str.replace(r"[\[\]\']", '', regex=True)

# Chia các thể loại thành các mục riêng lẻ và làm phẳng danh sách
genres_list = manga_data['Genres'].str.split(', ').explode()

# Đếm số lần xuất hiện của từng thể loại
genre_counts = genres_list.value_counts()

# Lấy 10 thể loại hàng đầu theo số lượng
top_10_genres = genre_counts.head(10)

# Vẽ biểu đồ cột cho 10 thể loại hàng đầu
plt.figure(figsize=(12, 6))

# Xác định bảng màu
colors = plt.cm.Paired.colors[:len(top_10_genres)]  # Sử dụng bản đồ màu 'Paired', cắt theo số lượng thể loại hàng đầu

# Tạo biểu đồ cột với
top_10_genres.plot(kind='bar', color=colors, edgecolor='black', linewidth=1.5)

# Tiêu đề và nhãn
plt.title('Top 10 Manga Genres by Count', fontsize=16, weight='bold')
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Điều chỉnh x-ticks để phù hợp với genres
plt.xticks(rotation=45, ha='right', fontsize=12)

# Hiện biểu đồ
plt.tight_layout()
plt.show()
```

```python
manga_copy_df = manga_df.copy()
manga_copy_df['Genres'] = manga_copy_df['Genres'].apply(eval)

# Phân tách tập dữ liệu để mỗi thể loại có một hàng riêng với các chi tiết về từng bộ truyện
exploded_data_by_genre = manga_copy_df.explode('Genres')

# Nhóm dữ liệu theo thể loại và tính toán các số liệu cần thiết cho từng thể loại
genre_grouped = exploded_data_by_genre.groupby('Genres').agg({
    'Favorite': 'mean',  # Average favorite count
    'Score': 'mean',     # Average score
    'Title': 'count',    # Number of manga
    'Popularity': 'mean' # Average popularity
}).reset_index()

# Tạo một cột điểm tổng hợp (điểm trung bình của điểm chuẩn hóa của bốn số liệu)
scaler = MinMaxScaler()
genre_grouped[['Favorite_norm', 'Score_norm', 'Title_norm', 'Popularity_norm']] = scaler.fit_transform(
    genre_grouped[['Favorite', 'Score', 'Title', 'Popularity']]
)

genre_grouped['Composite_Score'] = genre_grouped[['Favorite_norm', 'Score_norm', 'Title_norm', 'Popularity_norm']].mean(axis=1)

sorted_by_composite_score = genre_grouped.sort_values(by='Composite_Score', ascending=False)

top_genres_by_composite_score = sorted_by_composite_score.head(10)

# Trực quan hóa biểu đồ
palette = ["#11264e","#00507A","#026e90","#008b99","#6faea4","#fcdcb0","#FEE08B","#faa96e","#f36b3b","#ef3f28","#CC0028"]
plt.figure(figsize=(20,10))
p=sns.barplot(data=top_genres_by_composite_score,x='Composite_Score',y='Genres',palette=palette,saturation=1,edgecolor="#1c1c1c",linewidth=3)
plt.title('\nTop 10 Manga Genres by Composite Score\n',fontsize=30, color='k', weight='bold')
plt.xlabel('Composite Score',fontsize=20, weight='bold')
plt.ylabel('Genre',fontsize=20, weight='bold')

plt.yticks(fontsize=17)
plt.xticks(fontsize=17)
plt.xlim(min(top_genres_by_composite_score['Composite_Score']) - 0.1, max(top_genres_by_composite_score['Composite_Score']) + 0.1)
plt.show()
```

```python
df_copy = manga_df.copy()
df_copy['Author'] = df_copy['Author'].apply(literal_eval)

# Explode danh sách 'Tác giả' để tạo các hàng riêng biệt cho mỗi tác giả
df_exploded = df_copy.explode('Author')

# Nhóm theo tác giả để tính tổng số bộ truyện và điểm trung bình
author_df = df_exploded.groupby('Author').agg({'Title': 'count', 'Score': 'mean'}).reset_index()

# Đổi tên các cột để thống nhất
author_df.columns = ['Author', 'Total Manga', 'Average Score']

# Sắp xếp DataFrame theo 'Điểm trung bình' giảm dần
author_df = author_df.sort_values(by='Average Score', ascending=False)

# Trực quan biểu đồ 10 tác giả có điểm trung bình cao nhất
top_10_authors = author_df.head(10)
palette = ["#11264e","#00507A","#026e90","#008b99","#6faea4","#fcdcb0","#FEE08B","#faa96e","#f36b3b","#ef3f28","#CC0028"]
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_authors, x='Average Score', y='Author', palette=palette, linewidth=2, edgecolor='k')
plt.title('Top 10 Authors with the highest average score\n', fontsize=20, weight='bold')
plt.xlabel('Average Score', fontsize=15, weight='bold')
plt.ylabel('Author', fontsize=15, weight='bold')

plt.xlim(min(top_10_authors['Average Score']) - 0.1, max(top_10_authors['Average Score']) + 0.1)
plt.show()
```

**Nhận xét**
- Dựa vào biểu đồ ta thấy top 10 tác giả có điểm số trung bình cao nhất là `Yoshikawa Eiji`, `sing N song`, `Ai Qianshui de Wuzei`, `Mo Xiang Tong Xiu STARember`, `Abe Tsukasa`, `Shirabi Asato`, `Asato`, `Mikami Saka`, `Koyama Chuuya`, `Kim Huynso`
- Ta có thể thấy điểm số của 10 tác giả nằm khoảng 9 điểm.
- Dựa vào biểu đồ ta thấy 3 tác giả đứng đầu có điểm số vượt trội hơn so các tác giả còn lại

### 2.7. Câu hỏi 7:

**Câu hỏi:**
- Liệu một bộ truyện có điểm số trung bình cao có độ phổ biến lớn hơn không?
 
**Mục đích:**
- Phân tích mối quan hệ giữa **chất lượng đánh giá** (điểm số trung bình) và **sự yêu thích của độc giả** (độ phổ biến), nhằm xác định liệu bộ truyện có điểm số cao thường có độ phổ biến lớn hơn, hay có sự khác biệt nào trong cách mà người đọc đánh giá các bộ truyện.

**Ý tưởng**
- **Chọn các cột liên quan**: Bạn sẽ cần sử dụng các cột **`Score`** (điểm số trung bình) và **`Popularity`** (độ phổ biến) từ dữ liệu của mình.
- **Kiểm tra sự phân bố của dữ liệu**: Bạn có thể xem phân phối của điểm số trung bình và độ phổ biến để nhận diện những xu hướng cơ bản.
- **Phân tích mối quan hệ giữa điểm số và độ phổ biến**:
   - Tính toán hệ số tương quan (correlation) giữa **`Score`** và **`Popularity`** để xác định mức độ liên quan giữa chúng.
   - Vẽ biểu đồ **Scatter Plot** để trực quan hóa mối quan hệ giữa điểm số trung bình và độ phổ biến.


```python
# Tạo bản sao của DataFrame 
df = manga_df.copy()

# Bước 1: Kiểm tra mối quan hệ giữa điểm số và độ phổ biến
# Tính hệ số tương quan giữa điểm số và độ phổ biến
correlation = df['Score'].corr(df['Popularity'])
print(f'Correlation coefficient between score and popularity: {correlation}')

# Bước 2: Vẽ biểu đồ phân tán
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Score', y='Popularity', color='blue')

# Cài đặt tiêu đề và các nhãn
plt.title('Relationship between Manga Score and Popularity')
plt.xlabel('Average score')
plt.ylabel('Popularity')

# Ghi hệ số tương quan gần gốc tọa độ
plt.text(x=df['Score'].min() + 0.5,  # Gần giá trị nhỏ nhất của Score + một khoảng nhỏ
         y=df['Popularity'].min() - 0.05 * (df['Popularity'].max() - df['Popularity'].min()),  # Gần giá trị nhỏ nhất của Popularity
         s=f"Hệ số tương quan: {correlation:.2f}",
         fontsize=12, color='red', ha='left', bbox=dict(facecolor='white', alpha=0.5))

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

**Nhận xét**
1. **Mối quan hệ tương quan âm (negative correlation)**: Hệ số tương quan gần -0.5 cho thấy một **mối quan hệ âm trung bình** giữa điểm số và độ phổ biến. Điều này có nghĩa là, khi điểm số của bộ truyện cao hơn thì độ phổ biến có xu hướng thấp hơn, và ngược lại. Tuy nhiên, mối quan hệ này không phải là quá mạnh, vì hệ số này không gần với -1.

2. **Khả năng giải thích**:
   - Các bộ truyện có điểm số cao có thể không nhất thiết thu hút nhiều độc giả. Điều này có thể là do một số truyện có điểm số cao nhưng chỉ được một nhóm nhỏ độc giả đánh giá cao, chẳng hạn như những bộ truyện dành cho đối tượng cụ thể hoặc thể loại niche.
   - Ngược lại, những truyện có độ phổ biến cao có thể không được đánh giá cao bởi tất cả người xem, có thể là do một số yếu tố như sự quảng bá mạnh mẽ hoặc là các bộ truyện dễ tiếp cận với số đông người đọc.

### 2.8. Câu hỏi 8:

**Câu hỏi:**
- Các nhóm đối tượng có sự yêu thích khác nhau đối với các thể loại truyện nào?

**Mục đích:**
- Phân tích sở thích của nhóm đối tượng theo thể loại để xác định thể loại nào phổ biến với từng nhóm người.
- Từ đó giúp các nhà xuất bản truyện và các tác giả lập ra các chiến lược phát triển truyện nhắm đến từng nhóm đối tượng riêng.


**Ý tưởng:**
- Tách dữ liệu thể loại: Nếu các thể loại trong cột Genres có thể chứa nhiều thể loại (được phân tách bằng dấu phẩy hoặc dấu phân cách khác), ta cần tách chúng ra thành các giá trị riêng biệt.
- Tạo bảng thống kê: Tính tổng số lượt yêu thích (hoặc một chỉ số yêu thích khác) theo mỗi nhóm đối tượng và thể loại.
- Trực quan hóa dữ liệu: Sử dụng biểu đồ cột để thể hiện sự phân bố sở thích của các nhóm đối tượng với các thể loại khác nhau.

```python
# Tạo bản sao của DataFrame
df = manga_df.copy()

# Bước 1: Tách các thể loại truyện thành các thể loại riêng biệt (nếu có nhiều thể loại trong mỗi dòng)
df['Genres'] = df['Genres'].apply(lambda x: x.replace("'", "").replace("[", "").replace("]", "").split(", ") if isinstance(x, str) else [])

# Bước 2: Tạo một danh sách các thể loại với nhóm đối tượng
genre_demographics = []
for _, row in df.iterrows():
    for genre in row['Genres']:
        genre_demographics.append({'Genre': genre, 'Demographics': row['Demographic'], 'Favorite': row['Favorite']})

# Chuyển đổi thành DataFrame
genre_demographics_df = pd.DataFrame(genre_demographics)

# Bước 3: Tính tổng số lượt yêu thích theo thể loại và nhóm đối tượng
genre_demographics_summary = genre_demographics_df.groupby(['Genre', 'Demographics']).agg(
    total_favorites=('Favorite', 'sum'),
    count=('Favorite', 'count')
).reset_index()

# Bước 4: Vẽ biểu đồ
# Lọc dữ liệu theo thể loại phổ biến (nếu cần thiết)
top_genres = genre_demographics_summary.groupby('Genre').agg(
    total_favorites=('total_favorites', 'sum')
).sort_values(by='total_favorites', ascending=False).head(10).index

# Lọc dữ liệu chỉ chứa các thể loại phổ biến
genre_demographics_summary_filtered = genre_demographics_summary[genre_demographics_summary['Genre'].isin(top_genres)]

# Vẽ biểu đồ cột để phân tích sự yêu thích theo nhóm đối tượng
plt.figure(figsize=(12, 8))

# Tạo biểu đồ cột cho mỗi thể loại
for genre in top_genres:
    genre_data = genre_demographics_summary_filtered[genre_demographics_summary_filtered['Genre'] == genre]
    plt.bar(genre_data['Demographics'], genre_data['total_favorites'], label=genre)

plt.title('Preferences of target groups by manga genre')
plt.xlabel('Demographic group')
plt.ylabel('Total number of favourites')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

**Nhận xét**

1. **Thể loại và nhóm đối tượng nổi bật:**
   - **Nhóm đối tượng Shounen (dành cho thiếu niên nam)** là nhóm có tổng số lượt yêu thích lớn nhất so với các nhóm đối tượng khác. Các thể loại như "Action", "Fantasy", "Psychological", "Romance" và "Adventure" dường như rất phổ biến trong nhóm này.
   - **Nhóm Seinen (dành cho nam trưởng thành)** cũng có số lượt yêu thích đáng kể, với sự tập trung vào các thể loại như "Drama" và "Psychological".
    - Nhóm **Shoujo (dành cho thiếu niên nữ)** dành sự quan tâm nhiều nhất cho thể loại "Romance"
2. **Sự chênh lệch giữa các nhóm đối tượng:**
   - Nhóm **Shounen** vượt trội rõ rệt so với các nhóm khác, thể hiện rằng đây là nhóm độc giả lớn nhất hoặc có xu hướng bày tỏ sự yêu thích qua các nền tảng.
   - Nhóm **Shoujo (dành cho thiếu niên nữ)** có số lượt yêu thích thấp hơn nhiều so với Shounen và Seinen. Điều này có thể phản ánh mức độ quan tâm hoặc sự phổ biến của các thể loại trong nhóm này thấp hơn.
   - Nhóm **Josei (dành cho nữ trưởng thành)** và **Kids (trẻ em)** có số lượt yêu thích rất thấp, và dường như không có thể loại nào thực sự nổi bật.
3. **Sự đa dạng về thể loại trong các nhóm đối tượng:**
   - Nhóm Shounen và Seinen (đặc biệt là Shounen)có sự đa dạng cao về thể loại yêu thích (nhiều thể loại xuất hiện với số lượng tương đối đồng đều).
   - Các nhóm còn lại không có sự đa dạng về thể loại yêu thích
4. **Các thể loại phổ biến nhất:**
   - "Action" và "Fantasy" là hai thể loại nổi bật nhất, đặc biệt trong nhóm Shounen, cho thấy đây là những thể loại được yêu thích rộng rãi bởi độc giả trẻ.
   - "Drama" và "Psychological" có xu hướng phổ biến hơn trong nhóm Seinen, phù hợp với độc giả trưởng thành thích những câu chuyện có chiều sâu.
